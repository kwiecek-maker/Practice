```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 10, 0.01)
f_x = x
f_exp = np.exp(x)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2)

ax[0].plot(x, f_x)
ax[0].plot(x, f_exp)
ax[0].set_title('y-axis linear scale')

ax[1].plot(x, f_x)
ax[1].plot(x, f_exp)
ax[1].set_yscale('log')
ax[1].set_title('y-axis log scale')

plt.show()
```
