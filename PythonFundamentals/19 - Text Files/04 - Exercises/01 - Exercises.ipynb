````python
header = "i,fibonacci_i,factorial_i,gcd_fibonacci_i_factorial_i\n"
fib = []
with open("math_series.txt", "w") as f:
    f.write(header)
    for i in range(100):
        if i > 1:
            fib.append(fib[-1] + fib[-2])
        else:
            fib.append(1)
        curr_fib = fib[-1]
        factor = factorial(i)
        f.write(f"{i},{curr_fib},{factor},{gcd(curr_fib, factor)}\n")    
````
